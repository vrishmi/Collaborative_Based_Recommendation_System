```python
import pandas as pd
user_df=pd.read_csv('/content/user.csv', delimiter='\t')
user_df.head(15)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(user_df.info())
```

```python
print(user_df.describe())
```

```python
print(user_df.isnull().sum())
```

```python
# Check unique values in the 'clubs' column
print(user_df['clubs'].unique())

# Visualize the distribution of 'clubs' (top 10)
plt.figure(figsize=(12, 6))
sns.countplot(x='clubs', data=user_df, palette='viridis', order=user_df['clubs'].value_counts().index[:10])
plt.title('Top 10 User Clubs Distribution')
plt.xlabel('Club')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
user_df['last_online_date'] = pd.to_datetime(user_df['last_online_date'])
```

```python
for col in user_df.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(user_df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Drop non-numeric columns before calculating correlation
numeric_df = user_df.select_dtypes(include=['int64', 'float64'])

# Calculate correlation on the numeric DataFrame
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Check the column names in your DataFrame
print(user_df.columns)

# Time Series Analysis
user_df['last_online_date'] = pd.to_datetime(user_df['last_online_date'])
user_df.set_index('last_online_date', inplace=True)
user_df.resample('M').size().plot()
plt.title('Monthly User Activity')
plt.xlabel('Date')
plt.ylabel('Number of Users')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(user_df['mean_score'], bins=30, kde=True, color='green')
plt.title('Distribution of Mean Score Given by Users')
plt.xlabel('Mean Score')
plt.ylabel('Frequency')
plt.show()
```

```python
numerical_columns = ['num_watching', 'num_completed', 'num_on_hold', 'num_dropped', 'num_plan_to_watch', 'num_days', 'mean_score']

# Plot histograms for numerical columns
plt.figure(figsize=(16, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(user_df[col], bins=30, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Plot box plots for numerical columns
plt.figure(figsize=(16, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=user_df[col])
    plt.title(f'Box Plot of {col}')
    plt.xlabel(col)

plt.tight_layout()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.histplot(anime_df['popularity_rank'].dropna(), bins=20, kde=True, color='orange')
plt.title('Distribution of Anime Popularity Ranks')
plt.xlabel('Popularity Rank')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y=anime_df['type'])
plt.title('Count of Anime Types')
plt.xlabel('Count')
plt.ylabel('Anime Type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y=anime_df['status'])
plt.title('Count of Anime Status')
plt.xlabel('Count')
plt.ylabel('Anime Status')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='score', y='type', data=anime_df, ci=None, palette='viridis')
plt.title('Average Scores for Different Anime Types')
plt.xlabel('Average Score')
plt.ylabel('Anime Type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(anime_df['num_episodes'].value_counts().head(50).index, bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Number of Episodes')
plt.xlabel('Number of Episodes')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(anime_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
genres = anime_df['genres'].str.split(',').explode().str.strip()
plt.figure(figsize=(16, 10))
sns.countplot(y=genres, order=genres.value_counts().head(10).index, palette='pastel')
plt.title('Distribution of Anime Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
# Relationship between number of episodes and score
plt.figure(figsize=(12, 6))
sns.scatterplot(x='num_episodes', y='score', data=anime_df, color='teal', alpha=0.5)
plt.title('Relationship between Number of Episodes and Score')
plt.xlabel('Number of Episodes')
plt.ylabel('Score')
plt.show()

# Top studios producing anime
top_studios = anime_df['studios'].str.split(',').explode().str.strip().value_counts().head(10)
plt.figure(figsize=(12, 6))
top_studios.plot(kind='bar', color='orange')
plt.title('Top 10 Anime Studios')
plt.xlabel('Studio')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Distribution of anime genres for a specific anime type (e.g., TV)
tv_anime_genres = anime_df[anime_df['type'] == 'TV']['genres'].str.split(',').explode().str.strip().value_counts().head(10)
plt.figure(figsize=(12, 6))
tv_anime_genres.plot(kind='bar', color='purple')
plt.title('Top 10 Genres for TV Anime')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Distribution of user scores for specific anime types
plt.figure(figsize=(12, 6))
sns.boxplot(x='type', y='score', data=anime_df, palette='husl')
plt.title('Distribution of User Scores for Anime Types')
plt.xlabel('Anime Type')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='popularity_rank', y='score', data=anime_df, color='green', alpha=0.5)
plt.title('Relationship between Popularity Rank and User Score')
plt.xlabel('Popularity Rank')
plt.ylabel('Score')
plt.show()
```

```python
# Relationship between members count and user score
plt.figure(figsize=(12, 6))
sns.scatterplot(x='members_count', y='score', data=anime_df, color='coral', alpha=0.5)
plt.title('Relationship between Members Count and User Score')
plt.xlabel('Members Count')
plt.ylabel('Score')
plt.show()
```

```python
# Relationship between watching count and user score
plt.figure(figsize=(12, 6))
sns.scatterplot(x='watching_count', y='score', data=anime_df, color='gold', alpha=0.5)
plt.title('Relationship between Watching Count and User Score')
plt.xlabel('Watching Count')
plt.ylabel('Score')
plt.show()
```

```python
# Distribution of source types
plt.figure(figsize=(10, 5))
sns.countplot(x='source_type', data=anime_df, palette='muted')
plt.title('Distribution of Anime Source Types')
plt.xlabel('Source Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
user_user_df = pd.read_csv('/content/user_user.csv',delimiter='\t')

user_user_df.head()
```

```python
user_user_df.info()

# Check for missing values
missing_values = user_user_df.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
# Number of unique users and friendships
unique_users_count = user_user_df[['userA', 'userB']].nunique().sum()
unique_friendships_count = user_user_df['friends'].sum()
print(f"Number of unique users: {unique_users_count}")
print(f"Number of friendships: {unique_friendships_count}")
```

```python
# Friendship trends over time
user_user_df['friendship_date'] = pd.to_datetime(user_user_df['friendship_date'])
user_user_df['year'] = user_user_df['friendship_date'].dt.year

plt.figure(figsize=(12, 6))
sns.countplot(x='year', hue='friends', data=user_user_df, palette='Set2')
plt.title('Friendship Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Friends (1: Yes, 0: No)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

anime_anime_df = pd.read_csv('/content/anime_anime.csv',delimiter='\t')
anime_anime_df.head()
```

```python
# Display basic information about the dataset
print(anime_anime_df.info())

# Check for missing values
print(anime_anime_df.isnull().sum())
```

```python
# Exploration of relationships between pairs of animes
plt.figure(figsize=(12, 6))
sns.countplot(x='recommendation', data=anime_anime_df, palette='viridis')
plt.title('Distribution of Recommendations')
plt.xlabel('Recommendation (0: No, 1: Yes)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='related', data=anime_anime_df, palette='muted')
plt.title('Distribution of Related Animes')
plt.xlabel('Related (0: No, 1: Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Relationship type distribution for related animes
plt.figure(figsize=(14, 8))
sns.countplot(x='relation_type', data=anime_anime_df[anime_anime_df['related'] == 1], palette='pastel')
plt.title('Distribution of Relationship Types for Related Animes')
plt.xlabel('Relation Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Distribution of relationships and recommendation URLs
plt.figure(figsize=(14, 8))
sns.countplot(x='relation_type', hue='recommendation', data=anime_anime_df, palette='Set2')
plt.title('Distribution of Relationship Types with Recommendation Status')
plt.xlabel('Relation Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Recommendation', loc='upper right', labels=['No', 'Yes'])
plt.show()
```
