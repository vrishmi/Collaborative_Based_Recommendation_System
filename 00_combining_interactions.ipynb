```python
import pandas as pd
import fastparquet
pd.set_option('display.max_columns', None)
```

```python
cols_to_read = [
    "user_id",
    "anime_id",
    "favorite",
    "review_score",
    "review_story_score",
    "review_animation_score",
    "review_sound_score",
    "review_character_score",
    "review_enjoyment_score",
    "score",
]
interactions = []
for i in range(70):
    fname = f"./user_anime/user_anime0000000000{i:>02}.csv"
    print(f"Reading {fname}")
    interactions.append(pd.read_csv(fname, delimiter="\t", usecols= cols_to_read))
    print(f"Read    {fname}")
    print("---"*30)
```

```python
interactions[0]
```

```python
intMat = pd.concat(interactions).reset_index(drop=True)
```

```python
intMat
```

```python
intMat.describe()
```

```python
intMat2 = intMat.drop(columns=["review_score","review_story_score","review_animation_score","review_sound_score","review_character_score","review_enjoyment_score"])
```

```python
intMat2.to_parquet("user_anime.parquet", engine="auto")
```

```python
intMat2 = pd.read_parquet("user_anime.parquet", engine="fastparquet")
```

```python
valid_scores = intMat2[~intMat2['score'].isna()]
invalid_scores = intMat2[intMat2['score'].isna()]
```
