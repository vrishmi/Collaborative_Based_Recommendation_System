```python
import matplotlib.pyplot  as plt
import pandas as pd
```

```python
bl_als = pd.read_json("./results/Baseline_ALS_gridCV.json")
bl_sgd = pd.read_json("./results/Baseline_SGD_gridCV.json")
knn_b_ii = pd.read_json("./results/KNNBasic_ii_gridCV.json")
knn_b_uu = pd.read_json("./results/KNNBasic_uu_gridCV.json")
knn_m_ii = pd.read_json("./results/KNNMean_ii_gridCV.json")
knn_m_uu = pd.read_json("./results/KNNMean_uu_gridCV.json")
knn_z_ii = pd.read_json("./results/KNNZscore_ii_gridCV.json")
knn_z_uu = pd.read_json("./results/KNNZscore_uu_gridCV.json")
knn_cm_ii = pd.read_json("./results/KNN_M_custom_ii_gridCV.json")
knn_cm_uu = pd.read_json("./results/KNN_M_custom_uu_gridCV.json")
svd_b = pd.read_json("./results/SVD_Biased_gridCV.json")
svd_nb = pd.read_json("./results/SVD_NBiased_gridCV.json")
```

```python
cols = ["mean_test_mse", "mean_test_mae", "std_test_mse", "std_test_mae", "params"]
sort_order = ["mean_test_mse", "mean_test_mae", "std_test_mse", "std_test_mae"]
```

```python
bl_als = bl_als[cols]
bl_als = bl_als.sort_values(sort_order, ascending=True).reset_index(drop=True)

bl_sgd = bl_sgd[cols]
bl_sgd = bl_sgd.sort_values(sort_order, ascending=True).reset_index(drop=True)

knn_b_ii = knn_b_ii[cols]
knn_b_ii = knn_b_ii.sort_values(sort_order, ascending=True).reset_index(drop=True)

knn_b_uu = knn_b_uu[cols]
knn_b_uu = knn_b_uu.sort_values(sort_order, ascending=True).reset_index(drop=True)

knn_m_ii = knn_m_ii[cols]
knn_m_ii = knn_m_ii.sort_values(sort_order, ascending=True).reset_index(drop=True)

knn_m_uu = knn_m_uu[cols]
knn_m_uu = knn_m_uu.sort_values(sort_order, ascending=True).reset_index(drop=True)

knn_z_ii = knn_z_ii[cols]
knn_z_ii = knn_z_ii.sort_values(sort_order, ascending=True).reset_index(drop=True)

knn_z_uu = knn_z_uu[cols]
knn_z_uu = knn_z_uu.sort_values(sort_order, ascending=True).reset_index(drop=True)

knn_cm_ii = knn_cm_ii[cols]
knn_cm_ii = knn_cm_ii.sort_values(sort_order, ascending=True).reset_index(drop=True)

knn_cm_uu = knn_cm_uu[cols]
knn_cm_uu=knn_cm_uu.sort_values(sort_order, ascending=True).reset_index(drop=True)

svd_b = svd_b[cols]
svd_b = svd_b.sort_values(sort_order, ascending=True).reset_index(drop=True)

svd_nb = svd_nb[cols]
svd_nb = svd_nb.sort_values(sort_order, ascending=True).reset_index(drop=True)
```

```python
dfs = {
    "bl_als": bl_als.head(1),
    "bl_als": bl_als.head(1),
    "bl_sgd": bl_sgd.head(1),
    "knn_b_ii": knn_b_ii.head(1),
    "knn_b_uu": knn_b_uu.head(1),
    "knn_m_ii": knn_m_ii.head(1),
    "knn_m_uu": knn_m_uu.head(1),
    "knn_z_ii": knn_z_ii.head(1),
    "knn_z_uu": knn_z_uu.head(1),
    "knn_cm_ii": knn_cm_ii.head(1),
    "knn_cm_uu": knn_cm_uu.head(1),
    "svd_b": svd_b.head(1),
    "svd_nb": svd_nb.head(1),
}
```

```python
df = pd.concat(dfs.values()).assign(method=dfs.keys())
cols = ["method"] + list(df.columns[:-1])
df = df[cols]
df.sort_values("mean_test_mse", inplace= True)
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
bars = plt.bar(df["method"], df["mean_test_mse"])
plt.ylim(1.25, plt.ylim()[1])
plt.xticks(rotation=30)
plt.subplots_adjust(bottom=0.3) 
plt.xlabel("Method")
plt.ylabel("Mean Square Error")
plt.title("Bar Chart of MSE")
for bar in bars:
    height = bar.get_height()
    plt.gca().annotate('{:.3f}'.format(height),
                       xy=(bar.get_x() + bar.get_width() / 2, height),
                       xytext=(0, 3),  # 3 points vertical offset
                       textcoords="offset points",
                       ha='center', va='bottom', rotation = 30)
```


* ### We can observe that matrix factorization methods outperformed the neighborhood method by more than 4.35%
* ### One more interesting observation is that even the worst item-item based neighborhood (1.464) method outperformed the best user-user based neighborhood method (1.466) method. This is because we have more number of users than number of items and we rather prefer small number of quality neighbor than large number of loosely related neighbors.
* ### Significance weighing plays a significant role: Weighing the similarity between two entities (based on number of corated) squeezed more accuracy out of neighborhood method and pushed the MSE below 1.40
