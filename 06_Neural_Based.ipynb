**Latent Factors Learning by Deep Learning**

**Step 1:** 

Randomly initializing some parameters. These parameters will be a set of latent factors for each user and anime. We have used 5 for now. Because each user will have a set of these factors and each movie will have a set of these factors,

**Step 2** 

We Calculated our predictions, by taking the dot product of each anime with each user. If, for example, the first latent user factor represents how much the user likes action animes and the first latent anime factor represents if the anime has a lot of action or not, then the product of those will be particularly high if either the user likes action animes and the anime has a lot of action in it or the user doesn't like action animes and the anime doesn't have any action in it. On the other hand, if we have a mismatch (a user loves action animes but the anime doesn't belong to an action anime category, or the user doesn't like action animes and it is one), the product will be very low.

**Step 3** 

Calculating our loss. We have taken mean squared error to represent the accuracy of a prediction.

We can optimize our parameters (the latent factors) using stochastic gradient descent, such as to minimize the loss. At each step, the stochastic gradient descent optimizer will calculate the match between each anime and each user using the dot product, and will compare it to the actual rating that each user gave to each anime. It will then calculate the derivative of this value and will adjust the weights by multiplying this by the learning rate. After this, the loss can get better and better, and the recommendations can also get better and better.

For example, assuming the latent factors range between -1 and +1, with positive numbers indicating stronger matches and negative numbers weaker ones, and the categories are Action, SuperNatural,Comedy,Drama and Romance, then we could represent the anime Naruto: Shippuden  as:

```python
naruto_shippuden= np.array([0.98,0.88,0.91,-0.9,-0.8])
```

Here, for instance, we are scoring highly action as 0.98, very supernatural as 0.88,not very romantic as -0.9,very dramatic as 0.9 and very old as -0.9. We could represent a user who likes modern supernatural action movies as:

```python
user1 = np.array([0.98,0.88,-0.8,0.7,-0.6])
```

and now we can now calculate the similarity or match between this combination by doing the dot product and summing up:

```python
(user1*naruto_shippuden).sum()
```

Making our model learn the latent factors, by analyzing the existing relations between users and animes, our model can figure out itself the features that seem important or not known as embeddings. 

We have attributed to each of our users and each of our animes a random vector of a certain length (here, n_factors=5), and making them learnable parameters. That means that at each step, when we compute the loss by comparing our predictions to our targets, we are computing the gradients of the loss with respect to those embedding vectors and updating them with the rules of SGD.

At the beginning, these latent factors don't actually mean anything since we have chosen them randomly, but by the end of training they are pretty meaningful.

```python
n_users  = len(dls.classes['user_id'])
n_animes = len(dls.classes['title'])
n_factors = 5

user_factors = torch.randn(n_users, n_factors)
anime_factors = torch.randn(n_animes, n_factors)
```

Creating a new PyTorch DotProduct module inheriting from Module. PyTorch already provides a Module class, which provides some basic foundations.

When our module is called, PyTorch calls a method in our class called forward, and will pass along to that the parameters that are included in the call.The input of the model is a tensor of shape batch_size x 2, where the first column (x[:, 0]) contains the user IDs and the second column (x[:, 1]) contains the anime IDs. Also, We are using the embedding layers to represent our matrices of user and anime latent factors.

```python
class DotProduct(Module):
    def __init__(self, n_users, n_animes, n_factors):
        """
        Constructor for the DotProduct model.

        Parameters:
        - n_users (int): Number of users.
        - n_animes (int): Number of animes.
        - n_factors (int): Number of factors (embedding dimensions).
        """
        # Embedding layer for user factors
        self.user_factors = Embedding(n_users, n_factors)
        # Embedding layer for anime factors
        self.anime_factors = Embedding(n_animes, n_factors)
        
    def forward(self, x):
        """
        Forward pass of the DotProduct model.

        Parameters:
        - x (torch.Tensor): Input tensor containing user and anime indices.

        Returns:
        - torch.Tensor: Predicted ratings.
        """
        # Retrieving user embeddings based on user indices in the input tensor
        users = self.user_factors(x[:,0])
        # Retrieving anime embeddings based on anime indices in the input tensor
        animes = self.anime_factors(x[:,1])
        # Calculating the dot product of user and anime embeddings and summing along dimension 1
        return (users * animes).sum(dim=1)
```

It is not quite so easy to directly interpret the embedding matrices. There are just too many factors. So applying, principal component analysis (PCA). The animes look like these based on two of the strongest PCA components.

```python
# grouping the 'merged' DataFrame by the 'title' column and counting the number of scores for each anime.
g = merged.groupby('title')['score'].count()
#sorting the anime titles in descending order based on the count of scores and selecting the top 1000 anime titles.
top_animes = g.sort_values(ascending=False).index.values[:1000]
#converting the top anime titles to their corresponding indices in the data loaders used by the learner.
top_idxs = tensor([learn.dls.classes['title'].o2i[m] for m in top_animes])
#retrieveing the embedding weights (factors) for the top anime indices from the model. The .cpu().detach() part is used to move the data to the CPU and detach it from the computation graph.
anime_w = learn.model.anime_factors.weight[top_idxs].cpu().detach()
# applying PCA to reduce the dimensionality of the anime factors to 3 principal components.
anime_pca = anime_w.pca(3)
fac0, fac1, fac2 = anime_pca.t()
#selecting the first 100 anime titles and their corresponding values for the first and third principal components
idxs = list(range(100))
X = fac0[idxs]
Y = fac2[idxs]
plt.figure(figsize=(12, 12))
#creating a scatter plot with the first principal component on the x-axis and the third principal component on the y-axis.
plt.scatter(X, Y)
for i, x, y in zip(top_animes[idxs], X, Y):
    plt.text(x, y, i, color=np.random.rand(3) * 0.7, fontsize=11)
plt.show()
```

Showing the same logic for Users

## Anime Recommendation Using Cosine Similarity

Using cosine similarity to recommend the top 5 anime titles similar to "Death Parade." The recommendation is based on the cosine similarity of their embedding vectors in the learned collaborative filtering model.

```python
#Retrieving the embedding weights (factors) for all anime titles from the trained collaborative filtering model.
anime_factors = learn.model.anime_factors.weight
#Getting the index of the anime title "Death Parade" in the data loaders
idx_death_parade = dls.classes['title'].o2i['Death Parade']
#Calculating the cosine similarity between the embedding vector of "Death Parade" and all other anime titles.
distances = nn.CosineSimilarity(dim=1)(anime_factors, anime_factors[idx_death_parade][None])
#Identifying the indices of the top 5 anime titles with the highest cosine similarity, excluding the input anime itself.
idx_top_5 = distances.argsort(descending=True)[1:6]
top_5_animes = [dls.classes['title'][idx] for idx in idx_top_5]

print("Top 5 recommendations:")
for i, anime in enumerate(top_5_animes, 1):
    print(f"{i}. {anime}")
```

## User Similarity Recommendations Using Cosine Similarity

Utilizing cosine similarity to recommend the top 5 users similar to the user with ID '5034' in the collaborative filtering model.


```python
user_factors = learn.model.user_factors.weight
user_idx = dls.classes['user_id'].o2i['5034']  
distances = nn.CosineSimilarity(dim=1)(user_factors, user_factors[user_idx][None])
idx_top_5_users = distances.argsort(descending=True)[1:6]
top_5_users = [dls.classes['user_id'][idx] for idx in idx_top_5_users]

print("Top 5 users similar to the specified user:")
for i, user in enumerate(top_5_users, 1):
    print(f"{i}. {user}")
```
